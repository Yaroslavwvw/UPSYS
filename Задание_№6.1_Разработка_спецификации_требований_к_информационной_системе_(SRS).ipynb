# **Задание №6. Разработка спецификации требований к информационной системе (SRS)**

## **Введение**

**Спецификация требований к программному обеспечению (Software Requirements Specification, SRS)** является одним из основных документов в управлении проектами создания информационных систем. Этот документ служит связующим звеном между заказчиком и разработчиками, определяя четкое понимание того, что должна делать система, каким требованиям она должна соответствовать и как будет оцениваться успешность проекта.

Согласно статистике Standish Group, около 70% IT-проектов терпят неудачу из-за неправильно определенных или непонятых требований.

Грамотно составленная спецификация требований позволяет:
- Снизить риски проекта за счет раннего выявления противоречий и неясностей
- Обеспечить общее понимание целей между всеми заинтересованными сторонами
- Создать основу для планирования, разработки и тестирования системы
- Предотвратить "расползание границ проекта" (scope creep)
- Сформировать критерии приемки системы

#### **Теоретическая часть:**

**Спецификация требований к программному обеспечению (SRS)** — это документ, который полностью описывает ожидаемое поведение разрабатываемой системы и определяет все функциональные и нефункциональные требования, которым она должна соответствовать.

**Требование** — это условие или возможность, которой должна обладать система для решения проблемы пользователя или достижения цели, либо условие или возможность, которой должна обладать система для выполнения контракта, стандарта, спецификации или других формально налагаемых документов.

#### **Основные цели SRS:**

1. **Коммуникация** — обеспечить общее понимание между заказчиком, пользователями и разработчиками
2. **Основа для планирования** — служить базой для оценки сроков, стоимости и ресурсов
3. **Контракт** — определять обязательства разработчиков перед заказчиком
4. **Валидация** — служить основой для проверки соответствия системы ожиданиям
5. **Базовая линия** — фиксировать согласованные требования для управления изменениями

### **2. Типы требований**

#### **2.1. Функциональные требования**

**Функциональные требования** определяют конкретные функции, которые должна выполнять система, описывая поведение системы при определенных условиях и входных данных.

**Характеристики функциональных требований:**
- Описывают, **ЧТО** система должна делать
- Определяют действия системы в ответ на входные данные
- Специфицируют поведение системы в различных ситуациях
- Могут быть проверены путем выполнения соответствующих сценариев



##### **Общий пример — Система управления библиотекой:**



**ФТ-001: Регистрация читателя**
- **Описание:** Система должна позволять библиотекарю регистрировать нового читателя в системе, сохраняя его личные данные, контактную информацию и тип читательского билета.
- **Входные данные:** ФИО, дата рождения, адрес, телефон, email, тип читательского билета
- **Выходные данные:** Уникальный номер читательского билета, подтверждение регистрации
- **Приоритет:** Высокий
- **Критерий проверки:** Успешное создание записи читателя в базе данных с присвоением уникального идентификатора

**ФТ-002: Поиск книги**
- **Описание:** Система должна предоставлять возможность поиска книг по названию, автору, ISBN, жанру или ключевым словам.
- **Входные данные:** Поисковый запрос (текстовая строка), критерии поиска
- **Выходные данные:** Список книг, соответствующих критериям поиска, с информацией о доступности
- **Приоритет:** Критический
- **Критерий проверки:** Система находит все релевантные книги за время не более 2 секунд

**ФТ-003: Выдача книги**
- **Описание:** Система должна регистрировать выдачу книги читателю, проверяя наличие задолженностей, лимит одновременно выданных книг и доступность экземпляра.
- **Входные данные:** Номер читательского билета, ISBN книги
- **Выходные данные:** Квитанция о выдаче с датой возврата, обновление статуса книги
- **Бизнес-правила:**
  - Читатель может одновременно иметь не более 5 книг
  - Читатель с просроченными книгами или задолженностями не может получить новые книги
  - Срок выдачи составляет 14 дней для обычных читателей и 30 дней для студентов и преподавателей
- **Приоритет:** Критический
- **Критерий проверки:** Система корректно регистрирует выдачу, обновляет статус книги и рассчитывает дату возврата

#### **2.2. Нефункциональные требования**

**Нефункциональные требования** определяют свойства и ограничения системы, описывая **КАК** система должна выполнять свои функции.



**Категории нефункциональных требований:**

**а) Требования к производительности**

Определяют временные характеристики работы системы и требования к использованию ресурсов.

**Общий пример — Интернет-магазин:**

  - **НФТ-П-001: Время отклика**
    - Система должна отображать результаты поиска товаров за время не более 1 секунды при нормальной нагрузке (до 1000 одновременных пользователей)

  - **НФТ-П-002: Пропускная способность**
    - Система должна поддерживать одновременную работу не менее 5000 пользователей без деградации производительности

  - **НФТ-П-003: Использование ресурсов**
    - Мобильное приложение не должно занимать более 100 МБ оперативной памяти в режиме активного использования
    - Приложение не должно потреблять более 5% заряда батареи за час использования

**б) Требования к надежности**

Определяют способность системы функционировать без сбоев в течение определенного времени.

- **НФТ-Н-001: Доступность**
  - Система должна быть доступна не менее 99,5% времени в течение месяца (допустимое время простоя не более 3,6 часов в месяц)

- **НФТ-Н-002: Восстановление после сбоя**
  - В случае сбоя система должна автоматически восстанавливать работу в течение не более 5 минут
  - Данные пользовательских транзакций за последние 24 часа не должны быть потеряны

- **НФТ-Н-003: Обработка ошибок**
  - Система должна корректно обрабатывать все виды ошибок ввода данных, отображая понятные сообщения пользователю
  - При критических ошибках система должна сохранять лог с детальной информацией для последующего анализа

**в) Требования к безопасности**

Определяют меры защиты системы и данных от несанкционированного доступа.

- **НФТ-Б-001: Аутентификация**
  - Система должна требовать надежную аутентификацию пользователей (минимум 8 символов, включая буквы, цифры и специальные символы)
  - Система должна поддерживать двухфакторную аутентификацию для администраторов

- **НФТ-Б-002: Шифрование данных**
  - Все пароли должны храниться в зашифрованном виде с использованием алгоритма bcrypt или аналогичного
  - Передача данных между клиентом и сервером должна осуществляться по протоколу HTTPS с использованием TLS 1.2 или выше

- **НФТ-Б-003: Контроль доступа**
  - Система должна реализовывать ролевую модель доступа (RBAC)
  - Каждое действие пользователя должно проверяться на соответствие его правам доступа

**г) Требования к удобству использования (Usability)**

Определяют легкость освоения и использования системы.

- **НФТ-Ю-001: Обучаемость**
  - Новый пользователь должен иметь возможность выполнить основные операции (регистрация, поиск, заказ) в течение 5 минут без обращения к документации

- **НФТ-Ю-002: Интуитивность интерфейса**
  - Интерфейс системы должен соответствовать стандартным паттернам проектирования для данной платформы
  - Все интерактивные элементы должны иметь визуальную обратную связь при взаимодействии

- **НФТ-Ю-003: Доступность**
  - Интерфейс должен соответствовать стандарту WCAG 2.1 уровня AA для обеспечения доступности для людей с ограниченными возможностями

**д) Требования к совместимости и переносимости**

- **НФТ-С-001: Кроссплатформенность**
  - Мобильное приложение должно функционировать на iOS версии 13.0 и выше, а также Android версии 8.0 и выше
  - Веб-версия должна корректно отображаться в браузерах Chrome, Firefox, Safari и Edge последних трёх версий

- **НФТ-С-002: Интеграция**
  - Система должна предоставлять REST API для интеграции со сторонними системами
  - Система должна поддерживать экспорт данных в форматах CSV, JSON и XML

### **3. Характеристики качественных требований**

Хорошо сформулированное требование должно обладать следующими характеристиками (принцип **SMART + дополнительные критерии**):

#### **S - Specific (Конкретное)**

Требование должно быть четким и недвусмысленным.

**Плохо:** "Система должна работать быстро"

**Хорошо:** "Система должна отображать список товаров в течение 1 секунды после запроса пользователя при нагрузке до 1000 одновременных пользователей"

#### **M - Measurable (Измеримое)**

Требование должно иметь критерии, по которым можно проверить его выполнение.

**Плохо:** "Система должна быть надежной"

**Хорошо:** "Система должна иметь коэффициент доступности не менее 99,5% в месяц"

#### **A - Achievable (Достижимое)**

Требование должно быть реалистичным с учетом технических, временных и бюджетных ограничений.

**Плохо:** "Мобильное приложение должно работать на всех устройствах, выпущенных за последние 15 лет"

**Хорошо:** "Мобильное приложение должно работать на устройствах с Android 8.0+ и iOS 13.0+"

#### **R - Relevant (Релевантное)**

Требование должно соответствовать целям проекта и бизнес-потребностям.

#### **T - Testable (Проверяемое)**

Должна существовать методика проверки выполнения требования.

**Плохо:** "Интерфейс должен быть интуитивно понятным"

**Хорошо:** "90% новых пользователей должны успешно выполнить базовый сценарий использования (регистрация, поиск товара, оформление заказа) в течение 10 минут без обращения к справке"

#### **Дополнительные характеристики:**

**Unambiguous (Однозначное)**
Требование не должно допускать множественных интерпретаций.

**Complete (Полное)**
Требование должно содержать всю необходимую информацию для реализации.

**Consistent (Согласованное)**
Требование не должно противоречить другим требованиям.

**Traceable (Отслеживаемое)**
Требование должно иметь уникальный идентификатор и связь с источником (бизнес-цель, заинтересованная сторона).

### **4. Методы сбора требований**

#### **4.1. Интервью с заинтересованными сторонами**

Личные беседы с пользователями, заказчиками и другими стейкхолдерами для выявления их потребностей и ожиданий.

**Преимущества:**
- Глубокое понимание контекста и потребностей
- Возможность задавать уточняющие вопросы
- Выявление скрытых требований

**Недостатки:**
- Требует значительных временных затрат
- Риск субъективности и неполноты информации

#### **4.2. Анкетирование и опросы**

Сбор информации от большого числа пользователей с помощью структурированных вопросников.

**Преимущества:**
- Быстрый сбор данных от многих пользователей
- Возможность количественного анализа

**Недостатки:**
- Низкий процент отклика
- Отсутствие глубины и контекста

#### **4.3. Наблюдение за работой пользователей**

Изучение текущих рабочих процессов путем наблюдения за пользователями в их рабочей среде.

**Преимущества:**
- Выявление реальных потребностей, а не декларируемых
- Понимание контекста использования

#### **4.4. Анализ существующих систем**

Изучение аналогичных систем для выявления лучших практик и типовых функций.

#### **4.5. Прототипирование**

Создание макетов или прототипов интерфейса для обсуждения с пользователями.

**Преимущества:**
- Визуализация помогает пользователям лучше формулировать требования
- Раннее выявление проблем проектирования

#### **4.6. Мозговой штурм и воркшопы**

Групповые сессии с заинтересованными сторонами для генерации идей и достижения консенсуса.

### **5. Структура документа SRS**

Типовая структура спецификации требований согласно [IEEE 830](http://www.math.uaa.alaska.edu/~afkjm/cs401/IEEE830.pdf) и [ISO/IEC/IEEE 29148](https://cdn.standards.iteh.ai/samples/72089/62bb2ea1ef8b4f33a80d984f826267c1/ISO-IEC-IEEE-29148-2018.pdf):

```
1. ВВЕДЕНИЕ
   1.1. Цель документа
   1.2. Область применения
   1.3. Определения, акронимы и сокращения
   1.4. Ссылки
   1.5. Обзор документа

2. ОБЩЕЕ ОПИСАНИЕ
   2.1. Контекст системы
   2.2. Функции системы (высокоуровневое описание)
   2.3. Характеристики пользователей
   2.4. Ограничения
   2.5. Допущения и зависимости

3. СПЕЦИФИЧЕСКИЕ ТРЕБОВАНИЯ
   3.1. Функциональные требования
        3.1.1. Требования к функциональности модуля X
        3.1.2. Требования к функциональности модуля Y
        ...
   3.2. Нефункциональные требования
        3.2.1. Требования к производительности
        3.2.2. Требования к надежности
        3.2.3. Требования к безопасности
        3.2.4. Требования к удобству использования
        3.2.5. Требования к совместимости
        3.2.6. Требования к масштабируемости
   3.3. Требования к интерфейсам
        3.3.1. Пользовательские интерфейсы
        3.3.2. Программные интерфейсы
        3.3.3. Аппаратные интерфейсы
        3.3.4. Коммуникационные интерфейсы

4. МОДЕЛИ СИСТЕМЫ
   4.1. Диаграммы вариантов использования
   4.2. Диаграммы последовательности
   4.3. Модель данных

5. ТРЕБОВАНИЯ К ДОКУМЕНТАЦИИ
   5.1. Руководство пользователя
   5.2. Техническая документация
   5.3. Справочная система

6. КРИТЕРИИ ПРИЕМКИ
   6.1. Функциональная приемка
   6.2. Нефункциональная приемка

ПРИЛОЖЕНИЯ
   A. Глоссарий
   B. Модели анализа
   C. Список изменений
```

## **Примеры спецификаций требований**

### **Пример 1. IT-система управления проектами**

#### **1. ВВЕДЕНИЕ**

##### **1.1. Цель документа**

Данный документ определяет требования к системе управления проектами "ProjectHub", предназначенной для планирования, отслеживания и контроля выполнения проектов в IT-компаниях малого и среднего бизнеса.

##### **1.2. Область применения**

Система "ProjectHub" предназначена для использования командами от 5 до 100 человек, работающими над программными проектами различной сложности. Система обеспечит:
- Управление проектами и задачами
- Коллаборацию между членами команды
- Отслеживание прогресса и отчетность
- Интеграцию с системами контроля версий

##### **1.3. Определения и сокращения**

- **SRS** - Software Requirements Specification (Спецификация требований к программному обеспечению)
- **REST API** - Representational State Transfer Application Programming Interface
- **Задача** - единица работы в проекте, имеющая определенную цель и сроки выполнения
- **Спринт** - временной период фиксированной длительности для выполнения набора задач
- **Бэклог** - приоритизированный список задач проекта

##### **1.4. Пользовательские роли**

**Разработчик**
- **Описание:** Член команды, выполняющий задачи по разработке
- **Количество:** 3-20 на проект
- **Технический уровень:** Высокий
- **Основные задачи:**
  - Просмотр назначенных задач
  - Обновление статуса задач
  - Добавление комментариев и вложений
  - Отслеживание времени работы

**Руководитель проекта**
- **Описание:** Специалист, управляющий проектом и координирующий работу команды
- **Количество:** 1-2 на проект
- **Технический уровень:** Средний/Высокий
- **Основные задачи:**
  - Создание и настройка проектов
  - Планирование спринтов
  - Назначение задач членам команды
  - Мониторинг прогресса проекта
  - Генерация отчетов

**Администратор**
- **Описание:** Технический специалист, управляющий системой
- **Количество:** 1-2 на организацию
- **Технический уровень:** Высокий
- **Основные задачи:**
  - Управление пользователями и правами доступа
  - Настройка системы
  - Мониторинг производительности
  - Резервное копирование данных

#### **2. ФУНКЦИОНАЛЬНЫЕ ТРЕБОВАНИЯ**

##### **2.1. Управление проектами**

**ФТ-001: Создание проекта**
- **Описание:** Руководитель проекта должен иметь возможность создать новый проект, указав название, описание, дату начала и окончания, участников команды
- **Входные данные:**
  - Название проекта (строка, 3-100 символов)
  - Описание (текст, до 5000 символов)
  - Дата начала (дата)
  - Плановая дата окончания (дата)
  - Список участников (выбор из зарегистрированных пользователей)
- **Выходные данные:**
  - Уникальный идентификатор проекта
  - Подтверждение создания
  - Уведомления участникам команды
- **Бизнес-правила:**
  - Дата окончания должна быть позже даты начала
  - В проекте должен быть хотя бы один участник
  - Название проекта должно быть уникальным в рамках организации
- **Приоритет:** Критический
- **Критерий проверки:** Проект успешно создается в системе, все участники получают уведомление, проект отображается в списке проектов

**ФТ-002: Создание задачи**
- **Описание:** Руководитель проекта или разработчик должны иметь возможность создать новую задачу в рамках проекта
- **Входные данные:**
  - Заголовок задачи (строка, 3-200 символов)
  - Описание (текст, до 10000 символов)
  - Приоритет (низкий, средний, высокий, критический)
  - Исполнитель (опционально, выбор из участников проекта)
  - Плановое время выполнения (часы)
  - Теги (опционально, список строк)
- **Выходные данные:**
  - Уникальный идентификатор задачи
  - Автоматически присвоенный статус "К выполнению"
  - Дата создания
- **Бизнес-правила:**
  - Задача может быть назначена только участнику текущего проекта
  - Задача по умолчанию создается со статусом "К выполнению"
- **Приоритет:** Критический
- **Критерий проверки:** Задача создается с корректными данными, отображается в списке задач проекта, исполнитель получает уведомление

**ФТ-003: Изменение статуса задачи**
- **Описание:** Исполнитель задачи должен иметь возможность изменить статус задачи (К выполнению → В работе → На проверке → Выполнено)
- **Входные данные:**
  - Идентификатор задачи
  - Новый статус
  - Комментарий (опционально)
- **Выходные данные:**
  - Обновленный статус задачи
  - Временная метка изменения
  - Запись в истории задачи
- **Бизнес-правила:**
  - Переход между статусами должен следовать определенному workflow
  - Только исполнитель или руководитель проекта могут изменять статус
  - При переводе в статус "На проверке" должен быть назначен ревьюер
- **Приоритет:** Критический
- **Критерий проверки:** Статус задачи корректно обновляется, изменения отражаются в реальном времени, заинтересованные стороны получают уведомления

##### **2.2. Коллаборация**

**ФТ-004: Комментирование задач**
- **Описание:** Участники проекта должны иметь возможность добавлять комментарии к задачам для обсуждения деталей реализации
- **Входные данные:**
  - Идентификатор задачи
  - Текст комментария (до 5000 символов)
  - Упоминания пользователей (опционально)
  - Вложения (опционально, изображения, документы до 10 МБ)
- **Выходные данные:**
  - Добавленный комментарий с временной меткой
  - Уведомления упомянутым пользователям
- **Бизнес-правила:**
  - Комментировать могут только участники проекта
  - Автор комментария может редактировать его в течение 24 часов
  - Удалить комментарий может только автор или администратор
- **Приоритет:** Высокий
- **Критерий проверки:** Комментарий успешно добавляется, отображается в хронологическом порядке, упомянутые пользователи получают уведомления

**ФТ-005: Вложения к задачам**
- **Описание:** Пользователи должны иметь возможность прикреплять файлы к задачам (дизайн-макеты, документы, скриншоты)
- **Входные данные:**
  - Идентификатор задачи
  - Файл (форматы: jpg, png, pdf, docx, xlsx, размер до 10 МБ)
- **Выходные данные:**
  - Ссылка на загруженный файл
  - Миниатюра для изображений
- **Бизнес-правила:**
  - Разрешены только определенные типы файлов
  - Максимальный размер файла 10 МБ
  - К одной задаче можно прикрепить до 20 файлов
- **Приоритет:** Средний
- **Критерий проверки:** Файлы успешно загружаются, хранятся и отображаются в задаче

##### **2.3. Отчетность и аналитика**

**ФТ-006: Формирование отчета по проекту**
- **Описание:** Руководитель проекта должен иметь возможность сформировать отчет о текущем состоянии проекта
- **Входные данные:**
  - Идентификатор проекта
  - Период отчета (опционально, дата начала и окончания)
  - Метрики для включения в отчет (выбор из списка)
- **Выходные данные:**
  - PDF-документ с отчетом
  - Включает:
    - Общую информацию о проекте
    - Прогресс выполнения (% завершенных задач)
    - Распределение задач по статусам
    - Загрузка участников команды
    - Список просроченных задач
    - График burndown
- **Приоритет:** Высокий
- **Критерий проверки:** Отчет генерируется за время не более 5 секунд, содержит актуальные данные, корректно отображает графики

### **Пример 2. Мобильное приложение классификации болезней кожи по фото**

#### **1. ВВЕДЕНИЕ**

##### **1.1. Цель документа**

Настоящий документ определяет требования к мобильному приложению "SkinAI", предназначенному для предварительной классификации кожных заболеваний на основе фотографий с использованием технологий машинного обучения.

##### **1.2. Область применения**

Приложение "SkinAI" предназначено для обычных пользователей, беспокоящихся о состоянии своей кожи, и позволит:
- Получить предварительную информацию о возможном кожном заболевании по фотографии
- Ознакомиться с описанием выявленных состояний кожи
- Получить рекомендации о необходимости консультации с дерматологом
- Вести историю анализов для отслеживания динамики

**Важно:** Приложение НЕ является медицинским инструментом для постановки диагноза и НЕ заменяет консультацию с квалифицированным специалистом.

##### **1.3. Определения и сокращения**

- **ML** - Machine Learning (Машинное обучение)
- **CNN** - Convolutional Neural Network (Сверточная нейронная сеть)
- **Accuracy** - Точность классификации модели машинного обучения
- **Confidence score** - Показатель уверенности модели в результате классификации (от 0 до 1)
- **Дерматолог** - врач-специалист по заболеваниям кожи
- **Классификация** - процесс определения категории (типа заболевания) на основе характеристик изображения

##### **1.4. Пользовательские роли**

**Обычный пользователь**
- **Описание:** Человек без медицинского образования, обеспокоенный состоянием своей кожи
- **Возраст:** 18-65 лет
- **Технический уровень:** Низкий/Средний (умеет пользоваться смартфоном и базовыми приложениями)
- **Основные задачи:**
  - Загрузка или съемка фотографии кожи
  - Получение информации о возможном заболевании
  - Просмотр рекомендаций
  - Ведение истории анализов

**Медицинский эксперт (расширенная версия)**
- **Описание:** Дерматолог, использующий приложение как вспомогательный инструмент
- **Технический уровень:** Средний/Высокий
- **Основные задачи:**
  - Просмотр детальной статистики классификации
  - Доступ к научным ссылкам и исследованиям
  - Обратная связь по качеству классификации

#### **2. ФУНКЦИОНАЛЬНЫЕ ТРЕБОВАНИЯ**

##### **2.1. Регистрация и аутентификация**

**ФТ-001: Регистрация пользователя**
- **Описание:** Пользователь должен иметь возможность зарегистрироваться в приложении с использованием email или через социальные сети
- **Входные данные:**
  - Email (валидный email-адрес)
  - Пароль (минимум 8 символов, содержащий буквы и цифры)
  - Дата рождения (для подтверждения возраста 18+)
  - Согласие с условиями использования и политикой конфиденциальности
- **Альтернативный способ:** Регистрация через Google, Apple ID
- **Выходные данные:**
  - Уникальный идентификатор пользователя
  - Email с подтверждением регистрации
- **Бизнес-правила:**
  - Возраст пользователя должен быть не менее 18 лет
  - Email должен быть уникальным в системе
  - При регистрации через социальные сети email берется из профиля
- **Приоритет:** Критический
- **Критерий проверки:** Пользователь успешно создает аккаунт, получает подтверждение на email и может войти в приложение

**ФТ-002: Авторизация**
- **Описание:** Зарегистрированный пользователь должен иметь возможность войти в приложение
- **Входные данные:**
  - Email и пароль ИЛИ
  - Аутентификация через Google/Apple ID
- **Выходные данные:**
  - Доступ к функциям приложения
  - Токен сессии для последующих запросов
- **Бизнес-правила:**
  - После 5 неудачных попыток входа аккаунт блокируется на 15 минут
  - Токен сессии действителен 30 дней
- **Приоритет:** Критический
- **Критерий проверки:** Пользователь успешно входит с корректными данными, получает сообщение об ошибке при неправильных данных

##### **2.2. Получение и обработка изображений**

**ФТ-003: Съемка фотографии кожи**
- **Описание:** Пользователь должен иметь возможность сфотографировать проблемный участок кожи непосредственно в приложении
- **Входные данные:**
  - Доступ к камере устройства
  - Фотография проблемного участка кожи
- **Выходные данные:**
  - Изображение в формате JPEG
  - Метаданные: дата и время съемки, разрешение
- **Функциональность:**
  - Подсказки для правильного позиционирования камеры
  - Индикация достаточности освещения
  - Возможность использования вспышки
  - Предварительный просмотр и возможность пересъемки
- **Бизнес-правила:**
  - Минимальное разрешение фотографии 1280x720
  - Размер файла не более 10 МБ
- **Приоритет:** Критический
- **Критерий проверки:** Пользователь успешно делает фото, получает подсказки при недостаточном качестве, фото сохраняется для анализа

**ФТ-004: Загрузка существующего изображения**
- **Описание:** Пользователь должен иметь возможность выбрать изображение из галереи устройства
- **Входные данные:**
  - Доступ к галерее устройства
  - Выбор изображения в формате JPEG, PNG
- **Выходные данные:**
  - Загруженное изображение
  - Проверка качества изображения
- **Бизнес-правила:**
  - Поддерживаемые форматы: JPEG, PNG
  - Размер файла не более 10 МБ
  - Минимальное разрешение 640x480
- **Приоритет:** Критический
- **Критерий проверки:** Пользователь успешно выбирает и загружает изображение из галереи

**ФТ-005: Предварительная обработка изображения**
- **Описание:** Система должна автоматически обрабатывать загруженное изображение для повышения качества анализа
- **Входные данные:**
  - Исходное изображение кожи
- **Выходные данные:**
  - Обработанное изображение
  - Оценка качества изображения (приемлемо/неприемлемо)
- **Функциональность:**
  - Проверка качества изображения (резкость, освещенность, контрастность)
  - Автоматическое кадрирование релевантной области
  - Коррекция яркости и контраста при необходимости
  - Удаление шумов
  - Нормализация размера
- **Бизнес-правила:**
  - Изображение отклоняется, если:
    - Размытие превышает допустимый порог
    - Освещение слишком слабое или слишком яркое
    - Не обнаружена область кожи на изображении
- **Приоритет:** Критический
- **Критерий проверки:** Система корректно обрабатывает изображение, отклоняет изображения низкого качества с понятным объяснением причины

##### **2.3. Классификация заболеваний**

**ФТ-006: Анализ изображения и классификация**
- **Описание:** Система должна анализировать изображение кожи с помощью модели машинного обучения и определять возможное состояние кожи
- **Входные данные:**
  - Обработанное изображение кожи
  - Контекстная информация (опционально): часть тела, симптомы
- **Выходные данные:**
  - Основной результат классификации (наиболее вероятное состояние)
  - Показатель уверенности модели (confidence score, 0-100%)
  - Список альтернативных вариантов (до 3 наиболее вероятных состояний с показателями уверенности)
  - Время обработки
- **Функциональность:**
  - **В онлайн-режиме:**
    - Отправка изображения на сервер
    - Обработка с использованием полной модели CNN
    - Получение результата с высокой точностью
  - **В офлайн-режиме:**
    - Обработка с использованием облегченной модели на устройстве
    - Ограниченная точность (на 10-15% ниже)
    - Предупреждение о пониженной точности
- **Бизнес-правила:**
  - Модель должна классифицировать минимум 15 различных состояний кожи, включая:
    - Акне (угревая сыпь)
    - Атопический дерматит (экзема)
    - Псориаз
    - Розацеа
    - Контактный дерматит
    - Себорейный дерматит
    - Крапивница
    - Витилиго
    - Меланома (рак кожи)
    - Базалиома
    - Грибковые инфекции (микоз)
    - Герпес
    - Бородавки
    - Здоровая кожа
    - Неопределенное состояние
  - Если confidence score менее 40%, система должна отображать результат как "Неопределенное состояние"
  - При выявлении потенциально опасных состояний (меланома, базалиома) система должна обязательно рекомендовать немедленное обращение к врачу
- **Приоритет:** Критический
- **Критерий проверки:**
  - Модель классифицирует состояние за время не более 3 секунд в онлайн-режиме
  - Модель классифицирует состояние за время не более 1 секунды в офлайн-режиме
  - Точность классификации на тестовой выборке составляет не менее 85% (онлайн) и не менее 70% (офлайн)
  - Система корректно отображает результат с уровнем уверенности

##### **2.4. Отображение результатов и рекомендаций**

**ФТ-007: Отображение результатов классификации**
- **Описание:** Система должна понятно и структурированно представить результаты анализа пользователю
- **Входные данные:**
  - Результат классификации
  - Информация о состоянии кожи из базы знаний
- **Выходные данные:**
  - Экран с результатами, содержащий:
    - Название выявленного состояния кожи
    - Показатель уверенности модели
    - Краткое описание состояния
    - Типичные симптомы
    - Изображение для сравнения (эталонное фото состояния)
    - Уровень серьезности (низкий, средний, высокий, требует немедленного внимания)
    - Кнопки для перехода к подробной информации и рекомендациям
- **Бизнес-правила:**
  - Информация должна быть представлена на понятном, не медицинском языке
  - При низкой уверенности модели (менее 60%) должно отображаться предупреждение
  - При высокой серьезности состояния должен отображаться заметный предупредительный знак
- **Приоритет:** Критический
- **Критерий проверки:** Результаты отображаются понятно, структурированно, с визуальными элементами для лучшего восприятия

**ФТ-008: Подробная информация о заболевании**
- **Описание:** Пользователь должен иметь возможность просмотреть детальную информацию о выявленном состоянии кожи
- **Выходные данные:**
  - Развернутая информация:
    - Полное описание заболевания
    - Причины возникновения
    - Основные симптомы и проявления
    - Течение заболевания (острое, хроническое)
    - Возможные осложнения
    - Общая информация о лечении (без конкретных препаратов)
    - Меры профилактики
    - Когда необходимо обратиться к врачу
    - Ссылки на проверенные медицинские источники (опционально)
- **Приоритет:** Высокий
- **Критерий проверки:** Информация корректно отображается, является актуальной и понятной для неспециалиста

**ФТ-009: Персонализированные рекомендации**
- **Описание:** Система должна предоставлять рекомендации по дальнейшим действиям в зависимости от выявленного состояния и его серьезности
- **Входные данные:**
  - Результат классификации
  - Уровень серьезности состояния
  - Показатель уверенности модели
- **Выходные данные:**
  - Список рекомендаций с указанием срочности:
    - **Немедленно обратиться к врачу** (при подозрении на рак кожи, тяжелых инфекциях)
    - **Рекомендуется консультация в ближайшее время** (при хронических состояниях средней тяжести)
    - **Можно понаблюдать, при ухудшении обратиться к врачу** (при легких состояниях)
    - **Общие советы по уходу за кожей** (при здоровой коже или легких проблемах)
- **Функциональность:**
  - Ссылки на поиск ближайших дерматологических клиник (интеграция с картами)
  - Возможность добавить напоминание о записи к врачу
- **Приоритет:** Высокий
- **Критерий проверки:** Рекомендации соответствуют серьезности состояния, понятны и действенны

##### **2.5. История анализов**

**ФТ-010: Сохранение результатов анализа**
- **Описание:** Система должна автоматически сохранять все выполненные анализы в истории пользователя
- **Входные данные:**
  - Результат классификации
  - Исходное и обработанное изображение
  - Дата и время анализа
- **Выходные данные:**
  - Запись в истории пользователя
- **Бизнес-правила:**
  - История хранится локально на устройстве и синхронизируется с облаком
  - Максимальное количество записей в истории: 100
  - При превышении лимита удаляются самые старые записи
- **Приоритет:** Высокий
- **Критерий проверки:** Результаты успешно сохраняются и доступны в истории

**ФТ-011: Просмотр истории анализов**
- **Описание:** Пользователь должен иметь возможность просматривать все ранее выполненные анализы
- **Выходные данные:**
  - Список анализов, отсортированных по дате (новые сверху)
  - Для каждого анализа отображается:
    - Миниатюра изображения
    - Дата и время
    - Выявленное состояние
    - Показатель уверенности
- **Функциональность:**
  - Фильтрация по дате
  - Поиск по названию состояния
  - Удаление записей из истории
- **Приоритет:** Средний
- **Критерий проверки:** История корректно отображается, фильтрация и поиск работают без ошибок

**ФТ-012: Отслеживание динамики состояния**
- **Описание:** Пользователь должен иметь возможность отслеживать изменения состояния кожи на одном и том же участке с течением времени
- **Входные данные:**
  - Несколько анализов одного и того же участка кожи
- **Выходные данные:**
  - Сравнение изображений
  - График изменения показателей
  - Оценка динамики (улучшение, без изменений, ухудшение)
- **Приоритет:** Низкий
- **Критерий проверки:** Система корректно группирует анализы одного участка, визуализирует динамику

#### **3. НЕФУНКЦИОНАЛЬНЫЕ ТРЕБОВАНИЯ**

##### **3.1. Требования к производительности**

**НФТ-П-001: Время классификации**
- **Онлайн-режим:** Анализ изображения и получение результата должны занимать не более 5 секунд от момента отправки фото до отображения результата (при стабильном интернет-соединении)
- **Офлайн-режим:** Анализ изображения должен занимать не более 2 секунд

**НФТ-П-002: Время загрузки приложения**
- Начальная загрузка приложения (холодный старт) не должна превышать 3 секунд на устройствах среднего класса
- Повторная загрузка (теплый старт) не должна превышать 1 секунды

**НФТ-П-003: Использование ресурсов устройства**
- Приложение не должно занимать более 150 МБ пространства при установке
- Приложение не должно использовать более 200 МБ оперативной памяти в режиме активного использования
- Приложение не должно потреблять более 7% заряда батареи за час использования

**НФТ-П-004: Размер загружаемых данных**
- Обновление модели машинного обучения не должно превышать 50 МБ
- Загрузка базы знаний о заболеваниях не должна превышать 20 МБ

##### **3.2. Требования к надежности**

**НФТ-Н-001: Доступность сервиса**
- Серверная часть (API) должна быть доступна не менее 99% времени в месяц
- При недоступности сервера приложение должно переключаться в офлайн-режим с ограниченной функциональностью

**НФТ-Н-002: Обработка ошибок**
- Приложение не должно аварийно завершаться при возникновении ошибок
- Все ошибки должны обрабатываться корректно с отображением понятных сообщений пользователю
- При критических ошибках должны создаваться логи для последующего анализа

**НФТ-Н-003: Резервное копирование данных**
- История анализов должна автоматически синхронизироваться с облачным хранилищем при наличии интернета
- Пользователь должен иметь возможность восстановить свою историю при переустановке приложения или смене устройства

**НФТ-Н-004: Работа в нестабильных условиях**
- Приложение должно корректно работать при слабом интернет-соединении
- При потере соединения во время анализа приложение должно сохранять запрос и повторить попытку при восстановлении связи

##### **3.3. Требования к безопасности и конфиденциальности**

**НФТ-Б-001: Конфиденциальность данных**
- Все фотографии и медицинская информация пользователей должны храниться в зашифрованном виде
- Фотографии не должны передаваться третьим лицам без явного согласия пользователя
- Данные не должны использоваться для целей, не связанных с функциональностью приложения

**НФТ-Б-002: Передача данных**
- Вся передача данных между приложением и сервером должна осуществляться по защищенному протоколу HTTPS с TLS 1.3
- Изображения должны передаваться в зашифрованном виде

**НФТ-Б-003: Хранение данных на устройстве**
- Данные, хранящиеся локально на устройстве, должны быть зашифрованы с использованием встроенных средств операционной системы (Keychain для iOS, Keystore для Android)
- При удалении приложения все локальные данные должны быть полностью удалены

**НФТ-Б-004: Аутентификация**
- Пароли должны храниться в виде хэша с использованием алгоритма bcrypt
- Токены сессии должны обновляться каждые 30 дней
- При входе с нового устройства пользователь должен получать уведомление на email

**НФТ-Б-005: Соответствие нормативным требованиям**
- Приложение должно соответствовать требованиям GDPR (Общий регламент по защите данных ЕС) в части обработки персональных данных
- Приложение должно соответствовать требованиям Федерального закона № 152-ФЗ "О персональных данных" (РФ)
- Пользователь должен иметь возможность:
  - Запросить экспорт всех своих данных
  - Запросить удаление всех своих данных
  - Отозвать согласие на обработку данных

##### **3.4. Требования к удобству использования (Usability)**

**НФТ-Ю-001: Интуитивность интерфейса**
- 90% пользователей должны успешно выполнить первый анализ (от съемки/загрузки фото до просмотра результата) в течение 5 минут без обращения к инструкции
- Все основные функции должны быть доступны не более чем за 3 нажатия

**НФТ-Ю-002: Четкость и понятность информации**

- Процесс анализа должен сопровождаться индикатором прогресса
- При длительных операциях (более 2 секунд) должен отображаться индикатор загрузки

**НФТ-Ю-004: Помощь пользователю**

- При низком качестве изображения должны отображаться конкретные советы по улучшению

**НФТ-Ю-005: Доступность**

- Контрастность текста и фона должна быть не менее 4.5:1 (соответствие WCAG 2.1 AA)

##### **3.7. Специфические требования к ML-модели**

**НФТ-МО-001: Точность классификации**
- Модель должна иметь точность (accuracy) не менее 85% на тестовой выборке в онлайн-режиме
- Модель должна иметь точность не менее 70% в офлайн-режиме

**НФТ-МО-002: Чувствительность и специфичность**
- Для опасных состояний (меланома, базалиома) чувствительность (recall) должна быть не менее 95% (минимизация ложноотрицательных результатов)
- Для безопасных состояний специфичность должна быть не менее 80%

**НФТ-МО-003: Обновление модели**
- Модель должна обновляться не реже одного раза в квартал с учетом новых данных и улучшений алгоритмов
- Пользователь должен получать уведомление о доступности новой версии модели
- Обновление модели не должно требовать обновления приложения

**НФТ-МО-004: Прозрачность модели (Interpretability)**
- Для каждого результата классификации система должна предоставлять визуализацию областей изображения, на которые модель обратила особое внимание (attention maps/heatmaps)
- Это помогает пользователю понять, на основании каких признаков модель сделала вывод

# СПЕЦИФИКАЦИЯ ТРЕБОВАНИЙ К ПРОГРАММНОМУ ОБЕСПЕЧЕНИЮ
## 1. ВВЕДЕНИЕ

### 1.1. Цель документа

Данный документ описывает требования к программному прототипу геоинформационной системы (ГИС) для планирования съёмочных мероприятий спутниками дистанционного зондирования Земли (ДЗЗ).
Спецификация служит основой для разработки, тестирования и приемки прототипа в рамках дипломного проекта.

### 1.2. Область применения

Разрабатываемая система предназначена для:

анализа доступности спутниковых наблюдений для выбранного участка местности;

расчёта расписания пролётов спутников ДЗЗ на основе параметров орбит (TLE) и модели SGP4;

визуализации орбит, зон видимости и временных окон наблюдения в картографическом интерфейсе.

Система используется в учебно-исследовательском контексте (кафедра, лаборатория, студенты/магистранты), а также может применяться как вспомогательный инструмент при планировании съёмок.

### 1.3. Определения, акронимы и сокращения

ДЗЗ — дистанционное зондирование Земли.

TLE (Two-Line Element) — формат двухстрочного представления орбитальных параметров спутников.

SGP4 — стандартная математическая модель прогноза движения спутника по орбите на основе TLE.

AOI (Area of Interest) — область интереса, участок местности, для которого планируется наблюдение.

Пролёт — временной интервал, когда спутник может наблюдать выбранный участок.

ГИС — геоинформационная система, программный комплекс для работы с пространственными данными.

### 1.4. Пользовательские роли

Роль 1: Аналитик ДЗЗ / Студент

Описание: основной пользователь, выполняющий анализ доступности спутниковых наблюдений.

Количество: 1–3 человека.

Технический уровень: средний (знаком с основами ГИС и ДЗЗ).

Основные задачи:

выбор участка интереса на карте;

выбор спутников и источников TLE;

запуск расчёта пролётов и анализ результатов;

экспорт расписания наблюдений.

Роль 2: Преподаватель / Научный руководитель

Описание: проверяет корректность работы прототипа, использует систему для демонстрации принципов планирования съёмок.

Количество: 1–2 человека.

Технический уровень: средний/высокий.

Основные задачи:

запуск типовых сценариев;

оценка корректности расчётов;

использование результатов в учебном процессе.

Роль 3: Администратор системы (условная роль)

Описание: отвечает за первоначальную настройку системы (источники TLE, список спутников).

Количество: 1 человек (разработчик/дипломник).

Основные задачи:

конфигурация источников TLE;

наполнение базы характеристиками спутников;

техническая поддержка прототипа.

## 2. ФУНКЦИОНАЛЬНЫЕ ТРЕБОВАНИЯ
### 2.1. Управление областью интереса (AOI)

ФТ-001: Выбор точки или полигона AOI на карте

Описание: система должна позволять пользователю задавать область интереса на карте (точка или простой полигон).

Входные данные: взаимодействие пользователя с картой (щелчок мышью / набор вершин полигона).

Выходные данные: координаты AOI (широта/долгота точки или геометрия полигона).

Бизнес-правила:

Полигон должен быть геометрически корректным (не самопересекаться).

Минимальное число вершин полигона — 3.

Приоритет: Критический.

Критерий проверки: пользователь может задать точку и полигон; координаты сохраняются и используются в последующих расчётах.

ФТ-002: Редактирование и удаление AOI

Описание: пользователь должен иметь возможность изменять геометрию ранее заданной области интереса или удалять её.

Входные данные: выбор AOI, действия пользователя (перемещение точек, удаление).

Выходные данные: обновлённая геометрия AOI или пустой список AOI (при удалении).

Бизнес-правила:

После редактирования полигон должен оставаться валидным.

Приоритет: Высокий.

Критерий проверки: редактирование и удаление AOI выполняются без ошибок; обновлённая AOI учитывается в новых расчётах.

ФТ-003: Сохранение и загрузка AOI

Описание: система должна позволять сохранять заданные AOI под именем и в дальнейшем загружать их.

Входные данные: имя AOI, геометрия, команда «Сохранить» / «Загрузить».

Выходные данные: запись в локальном хранилище / восстановление AOI на карте.

Бизнес-правила:

Имя AOI должно быть уникально в рамках пользователя.

Приоритет: Средний.

Критерий проверки: сохранённая AOI доступна для повторного использования после перезапуска системы.

### 2.2. Работа с орбитальными данными и спутниками

ФТ-004: Загрузка TLE из открытых источников

Описание: система должна загружать актуальные TLE-данные для выбранных спутников ДЗЗ из открытых ресурсов (например, Celestrak).

Входные данные: список идентификаторов спутников / набор TLE-файлов.

Выходные данные: сохранённый локально перечень TLE-записей.

Бизнес-правила:

Для каждого спутника хранится минимум одна актуальная TLE-запись.

При повторной загрузке старые TLE могут быть помечены как архивные.

Приоритет: Критический.

Критерий проверки: система успешно загружает и отображает TLE для не менее 3 спутников ДЗЗ.

ФТ-005: Управление списком спутников

Описание: пользователь должен иметь возможность выбирать, какие спутники участвуют в расчёте (включать/исключать).

Входные данные: список доступных спутников, действия пользователя (флаг выбора).

Выходные данные: актуальный список активных спутников для расчёта.

Бизнес-правила:

Минимум один спутник должен быть выбран для запуска расчёта.

Приоритет: Высокий.

Критерий проверки: в расчётах учитываются только отмеченные пользователем спутники.

ФТ-006: Просмотр характеристик спутника

Описание: система должна отображать основные характеристики спутника ДЗЗ (название, тип, высота орбиты, наклонение, период, разрешение съёмки).

Входные данные: выбор спутника в интерфейсе.

Выходные данные: карточка с характеристиками.

Приоритет: Средний.

Критерий проверки: при выборе спутника отображается корректная информация из базы характеристик.

### 2.3. Расчёт пролётов и видимости

ФТ-007: Расчёт положения спутников по SGP4

Описание: система должна рассчитывать положение выбранных спутников во времени на основе TLE и модели SGP4.

Входные данные: TLE для каждого спутника, временной интервал, шаг расчёта.

Выходные данные: временной ряд координат спутников (широта, долгота, высота).

Бизнес-правила:

Шаг расчёта по умолчанию — не более 60 секунд.

Горизонт прогноза по умолчанию — до 7 суток от текущей даты.

Приоритет: Критический.

Критерий проверки: для заданного интервала времени система возвращает непрерывный временной ряд положения спутника, без ошибок вычисления.

ФТ-008: Определение окон видимости AOI

Описание: система должна определять интервалы времени, когда спутник может наблюдать заданную область интереса.

Входные данные: временной ряд положения спутника, геометрия AOI, критерий видимости (максимальное расстояние/угол над горизонтом).

Выходные данные: список пролётов (дата начала, дата окончания, максимальный угол/минимальное расстояние).

Бизнес-правила:

Пролётом считается интервал, в течение которого спутник удовлетворяет заданному порогу видимости (например, угол над горизонтом ≥ N градусов).

Минимальная длительность пролёта может задаваться (например, ≥ 30 секунд).

Приоритет: Критический.

Критерий проверки: для тестового AOI и выбранного спутника система определяет пролёты, совпадающие (с разумной погрешностью) с эталонными расчётами.

ФТ-009: Фильтрация и настройка критериев видимости

Описание: пользователь должен иметь возможность задавать пороги видимости (минимальный угол, минимальная длительность пролёта) и фильтровать результаты.

Входные данные: параметры фильтра (угол, длительность), список пролётов.

Выходные данные: отфильтрованный список пролётов.

Приоритет: Высокий.

Критерий проверки: при изменении порогов список пролётов обновляется и соответствует новым условиям.

### 2.4. Визуализация в ГИС-интерфейсе

ФТ-010: Отображение карты и спутниковых орбит

Описание: система должна отображать базовую карту (например, OpenStreetMap) и траектории движения спутников над поверхностью Земли.

Входные данные: координаты спутников, параметры визуализации, выбор спутника.

Выходные данные: картографическое представление орбит (линии/следы).

Приоритет: Высокий.

Критерий проверки: при выборе спутника на карте отображается его орбита за заданный временной интервал.

ФТ-011: Визуализация AOI и зон видимости

Описание: система должна отображать AOI и зоны видимости спутника над этим участком (проекция следа, точка над пролётом).

Входные данные: геометрия AOI, рассчитанные пролёты, положение спутника во времени.

Выходные данные: отображение AOI и соответствующих положений спутника на карте.

Приоритет: Высокий.

Критерий проверки: при выборе пролёта пользователь видит соответствующую позицию спутника относительно AOI.

ФТ-012: Отображение расписания пролётов в табличном виде

Описание: система должна выводить список пролётов в виде таблицы.

Входные данные: результаты расчёта пролётов.

Выходные данные: таблица: спутник, начало, конец, длительность, максимальный угол/минимальное расстояние.

Приоритет: Высокий.

Критерий проверки: таблица корректно обновляется после пересчёта или изменения фильтров.

### 2.5. Экспорт и сценарии использования

ФТ-013: Экспорт расписания пролётов

## 3. НЕФУНКЦИОНАЛЬНЫЕ ТРЕБОВАНИЯ
### 3.1. Требования к производительности

Расчёт пролётов для 1 AOI и до 5 спутников на горизонте прогноза 3 суток должен выполняться не более чем за 30 секунд на целевой машине (ноутбук среднего уровня).

Время реакции на базовые действия пользователя (перемещение/масштабирование карты, выбор спутника, открытие таблицы) не должно превышать 1 секунды.

НФТ-П-003: Объём обрабатываемых данных

Система должна поддерживать расчёт для не менее 10 спутников и нескольких AOI без критической деградации производительности.

### 3.2. Требования к надёжности и устойчивости

НФТ-Н-001: Обработка ошибок расчёта

При возникновении ошибок (некорректный TLE, выход за пределы модели) система не должна аварийно завершаться, а должна выводить понятное сообщение пользователю.

НФТ-Н-002: Сохранение пользовательских данных

Сохранённые AOI и сценарии не должны теряться при перезапуске программы (локальное хранилище / файл).

### 3.3. Требования к безопасности

(для дипломного прототипа — базовый уровень)

НФТ-Б-001: Ограничение на сетевые операции

Система должна загружать TLE только с явно указанного в настройках списка источников.

НФТ-Б-002: Хранение конфигурации

Конфигурационные данные (пути к файлам, URL источников) должны храниться в текстовом файле или настройках без критических секретов (пароли не используются).

### 3.4. Требования к удобству использования (Usability)

НФТ-Ю-001: Порог входа

Пользователь с базовыми знаниями ГИС должен освоить основные операции (выбор AOI, запуск расчёта, просмотр пролётов) за не более 15 минут по краткой инструкции.

НФТ-Ю-002: Интуитивность интерфейса

Основные функции (выбор AOI, выбор спутников, запуск расчёта) должны быть доступны из главного окна не более чем в 2 клика.

НФТ-Ю-003: Язык интерфейса

Интерфейс системы должен быть доступен минимум на русском языке.

### 3.5. Требования к совместимости и переносимости

НФТ-С-001: Платформа

Прототип должен корректно работать на ОС Windows 10+ или в современном браузере (Chrome/Firefox) — в зависимости от выбранной реализации (desktop/web).

НФТ-С-002: Разрешение экрана

Интерфейс должен корректно отображаться при разрешении экрана от 1366×768 и выше.

### 3.6. Требования к масштабируемости

НФТ-М-001: Масштабируемость по числу спутников

Архитектура системы должна обеспечивать независимость от среды выполнения и возможность масштабирования за счёт контейнеризации. Система должна корректно функционировать на основных операционных системах (Windows, Linux, macOS) и поддерживать развёртывание в Docker-контейнерах без модификации исходного кода.

## 4. ТРЕБОВАНИЯ К ИНТЕРФЕЙСАМ
### 4.1. Пользовательский интерфейс

## 5. ОГРАНИЧЕНИЯ И ДОПУЩЕНИЯ
### 5.1. Ограничения

ОГР-001: Учебный прототип
Система разрабатывается как учебно-исследовательский прототип в рамках дипломного проекта и не предназначена для оперативного управления реальными спутниковыми программами.

ОГР-002: Модель SGP4 и точность TLE
Точность расчётов ограничена точностью входных TLE и применимостью модели SGP4; высокоточные профессиональные расчёты выходят за рамки проекта.

ОГР-003: Отсутствие много пользовательского режима
Система рассчитана на работу одним пользователем за одним рабочим местом; многопользовательский доступ не реализуется.

### 5.2. Допущения

ДОП-001: Наличие доступа к открытым TLE
Предполагается, что на момент использования система имеет доступ к открытым TLE-файлам (через интернет либо заранее скачанным данным).

ДОП-002: Ограничение по типу орбит
Предполагается, что основная часть анализируемых спутников — низкоорбитальные (LEO) аппараты ДЗЗ; геостационарные и другие типы орбит в системе рассматриваются ограниченно или не рассматриваются.
