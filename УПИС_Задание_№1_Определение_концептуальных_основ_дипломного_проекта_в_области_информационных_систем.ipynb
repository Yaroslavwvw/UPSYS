#### **3. Методология использования**


Каким образом предполагается практическое применение разрабатываемого решения? Опишите основные сценарии взаимодействия пользователей с разрабатываемым решением.

Практическое использование решения предполагает работу с ГИС-интерфейсом:

    1.Пользователь выбирает интересующий участок местности (координаты или полигон на карте).

    2.Система загружает актуальные орбитальные параметры спутников из открытых источников.

    3.Выполняется прогноз движения спутников по модели SGP4.

    4.На карте отображаются траектории пролётов и временные окна наблюдения выбранного участка.

    5.Пользователь может сравнить различные аппараты (по разрешению, спектральным диапазонам, частоте повторения).

Таким образом, пользователь получает инструмент для планирования наблюдений, отбора оптимальных спутников и оценки временной доступности данных.
